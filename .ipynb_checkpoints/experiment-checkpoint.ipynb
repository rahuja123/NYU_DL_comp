```python
from matplotlib import pyplot as plt
plt.rcParams['figure.dpi'] = 96 # change dpi to make plots bigger

def show_image(img_path, title=None):
    print(img_path)
    img = Image.open(img_path)
    img = img.convert('RGB')
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')


show_image(images[1])
```
